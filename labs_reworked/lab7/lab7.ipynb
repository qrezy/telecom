```python
def discrete_fourier_transform(ys):
    N = len(ys)
    ts = npy.arange(N) / N
    freqs = npy.arange(N)
    args = npy.outer(ts, freqs)
    M = npy.exp(1j * npy.pi * 2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps
```

```python
def fast_fourier_transform(ys):
    N = len(ys)
    if N == 1:
        return ys

    He = fft(ys[::2])
    Ho = fft(ys[1::2])

    ns = npy.arange(N)
    W = npy.exp(-1j * npy.pi * 2 * ns / N)

    return npy.tile(He, 2) + W * npy.tile(Ho, 2)
```
